```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
df = df.replace({"Gender" : 1}, {"Gender": "male"})
df
```

```python
df = df.replace({"Gender": 0}, {"Gender": "female"})
df
```

```python
df.isnull().sum()
```

```python
df.boxplot(column="Heart rate")
```

```python
q1 = df["Heart rate"].quantile(0.25)
q3 = df["Heart rate"].quantile(0.75)
q1,q3
```

```python
IQR = q3 - q1
IQR
```

```python
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
lowerLimit, upperLimit
```

```python
df = df[(df["Heart rate"]>lowerLimit)&(df["Heart rate"]<upperLimit)]
df
```

```python
df.boxplot(column="Systolic blood pressure")
```

```python
q1 = df["Systolic blood pressure"].quantile(0.25)
q3 = df["Systolic blood pressure"].quantile(0.75)
IQR = q3 - q1
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
```

```python
df = df[(df["Systolic blood pressure"]>lowerLimit)&(df["Systolic blood pressure"]<upperLimit)]
df
```

```python
q1 = df["Diastolic blood pressure"].quantile(0.25)
q3 = df["Diastolic blood pressure"].quantile(0.75)
IQR = q3 - q1
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
df = df[(df["Diastolic blood pressure"]>lowerLimit)&(df["Diastolic blood pressure"]<upperLimit)]
df
```

```python
q1 = df["Blood sugar"].quantile(0.25)
q3 = df["Blood sugar"].quantile(0.75)
IQR = q3 - q1
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
df = df[(df["Blood sugar"]>lowerLimit)&(df["Blood sugar"]<upperLimit)]
df
```

```python
q1 = df["CK-MB"].quantile(0.25)
q3 = df["CK-MB"].quantile(0.75)
IQR = q3 - q1
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
df = df[(df["CK-MB"]>lowerLimit)&(df["CK-MB"]<upperLimit)]
df
```

```python
q1 = df["Troponin"].quantile(0.25)
q3 = df["Troponin"].quantile(0.75)
IQR = q3 - q1
lowerLimit = q1-1.5*IQR
upperLimit = q3+1.5*IQR
df = df[(df["Troponin"]>lowerLimit)&(df["Troponin"]<upperLimit)]
df
```
